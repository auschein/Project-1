```python
# Get your zws-id from zillow and put in a config file you can call
# then you Must run this prior
#!pip install pyzillow

import pandas as pd
from zillow_config import zwsid
from pyzillow.pyzillow import ZillowWrapper, GetDeepSearchResults

df = pd.read_csv("78702_Address_List4.csv")

address = []
zipcodes = []
for item in df["Property Address"]:
    address.append(item)
for z in df["Zip"]:
    zipcodes.append(z)
ref = list(zip(address, zipcodes))

print('Done')
```

```python
zillow_id = []
home_type = []
tax_year = []
tax_value = []
year_built = []
property_size = []
home_size = []
bathrooms = []
bedrooms = []
last_sold_date = []
last_sold_price = []
zestimate_amount = []
 
    
for x,y in ref:
    addresses = x
    zipcode = y
    try:
        zillow_data = ZillowWrapper(zwsid)
        deep_search_response = zillow_data.get_deep_search_results(addresses,zipcode)
        result = GetDeepSearchResults(deep_search_response)
        
        R1= result.zillow_id
        R2 = result.home_type
        R3 = result.tax_year
        R4 = result.tax_value
        R5 = result.year_built
        R6 = result.property_size
        R7 = result.home_size
        R8 = result.bathrooms
        R9 = result.bedrooms
        R10 = result.last_sold_date
        R11 = result.last_sold_price
        R12 = result.zestimate_amount
        
        zillow_id.append(R1)
        home_type.append(R2)
        tax_year.append(R3)
        tax_value.append(R4)
        year_built.append(R5)
        property_size.append(R6)
        home_size.append(R7)
        bathrooms.append(R8)
        bedrooms.append(R9)
        last_sold_date.append(R10)
        last_sold_price.append(R11)
        zestimate_amount.append(R12)
 
    except:
        zillow_id.append('NaN')
        home_type.append('NaN')
        tax_year.append('NaN')
        tax_value.append('NaN')
        year_built.append('NaN')
        property_size.append('NaN')
        home_size.append('NaN')
        bathrooms.append('NaN')
        bedrooms.append('NaN')
        last_sold_date.append('NaN')
        last_sold_price.append('NaN')
        zestimate_amount.append('NaN')
    
        
#This is going to take some time
print(len(zillow_id))
```

```python
z_df = pd.DataFrame ({"zillow_id" : zillow_id,
                      "home_type" : home_type,
                      "tax_year" : tax_year,
                      "tax_value" : tax_value,
                      "year_built" : year_built,
                      "property_size" : property_size,
                      "home_size" : home_size,
                      "bathrooms" : bathrooms,
                      "bedrooms" : bedrooms,
                      "last_sold_date" : last_sold_date,
                      "last_sold_price" : last_sold_price,
                      "zestimate_amount" : zestimate_amount,
                      "Property Address" : address})

#Merge Data Sets and Remove the ones missing zillow Data

combined_df = pd.merge(df,z_df, on="Property Address", how = "inner")
zipref = str(zipcodes[0])
killNaNs = combined_df[ combined_df['zillow_id'] == 'NaN' ].index
combined_df.drop(killNaNs , inplace=True)

combined_df.to_csv(f"{zipref}_Updated_List4.csv" , index=False, header=True)
captured_rows = str(len(combined_df))

print(f"You got {captured_rows} good records! Look.. ")
combined_df.head(5)
```
